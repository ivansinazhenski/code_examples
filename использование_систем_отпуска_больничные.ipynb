## Cоздание сводной

```python
befs_illness.fillna('-', inplace=True)
```

```python
%%time

res_df = pd.DataFrame()

for mrf in befs_illness[('МРФ','-')].value_counts().index:
    mrf_bi = befs_illness[befs_illness[('МРФ','-')] == mrf].copy()
    i = mrf
    for cl in mrf_bi.columns:
        if cl[1] == '-' or cl[1] == 'Всего заявок' or cl[1] == 'Доля использования WT':
            continue
        res_df.loc[i, 'МРФ'] =' Итог'
        res_df.loc[i, 'РФ'] =' Итог'
        res_df.loc[i, 'ЦФО'] = ' Итог'
        res_df.loc[i, 'Подразделение'] =' Итог'
        try:
            res_df.loc[i, cl[0]] = mrf_bi[cl].sum()/ mrf_bi[cl[0], 'Всего заявок'].sum()
        except ZeroDivisionError:
            res_df.loc[i, cl[0]] = '-'
            
            
    for fill in mrf_bi[('РФ','-')].value_counts().index:
        rf_bi = mrf_bi[mrf_bi[('РФ','-')] == fill].copy()
        i = mrf + fill
        for cl in rf_bi.columns:
            if cl[1] == '-' or cl[1] == 'Всего заявок' or cl[1] == 'Доля использования WT':
                continue
            res_df.loc[i, 'МРФ'] = mrf
            res_df.loc[i, 'РФ'] =' Итог'
            res_df.loc[i, 'ЦФО'] =' Итог'
            res_df.loc[i, 'Подразделение'] =' Итог'
            try:
                res_df.loc[i, cl[0]] = rf_bi[cl].sum()/ rf_bi[cl[0], 'Всего заявок'].sum()
            except ZeroDivisionError:
                res_df.loc[i, cl[0]] = '-'
                
                
        for cfo in rf_bi[('ЦФО', '-')].value_counts().index:
            cfo_bi = rf_bi[rf_bi[('ЦФО', '-')] == cfo].copy()
            i = mrf + fill + cfo
            for cl in cfo_bi.columns:
                if cl[1] == '-' or cl[1] == 'Всего заявок' or cl[1] == 'Доля использования WT':
                    continue
                res_df.loc[i, 'МРФ'] = mrf
                res_df.loc[i, 'РФ'] = fill
                res_df.loc[i, 'ЦФО'] = ' Итог'
                res_df.loc[i, 'Подразделение'] =' Итог'
                try:
                    res_df.loc[i, cl[0]] = cfo_bi[cl].sum()/ cfo_bi[cl[0], 'Всего заявок'].sum()
                except ZeroDivisionError:
                    res_df.loc[i, cl[0]] = '-'
                    
                    
            for p in cfo_bi[('Подразделение','-')].value_counts().index:
                p_bi = cfo_bi[cfo_bi[('Подразделение','-')] == p].copy()
                i = p + mrf +fill + cfo
                for cl in p_bi.columns:
                    if cl[1] == '-' or cl[1] == 'Всего заявок' or cl[1] == 'Доля использования WT':
                        continue
                    res_df.loc[i, 'МРФ'] = mrf
                    res_df.loc[i, 'РФ'] = fill
                    res_df.loc[i, 'ЦФО'] = cfo 
                    res_df.loc[i, 'Подразделение'] = p
                    try:
                        res_df.loc[i, cl[0]] = p_bi[cl].sum()/ p_bi[cl[0], 'Всего заявок'].sum()
                    except ZeroDivisionError:
                        res_df.loc[i, cl[0]] = '-'
```
